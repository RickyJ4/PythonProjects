# **SEABORN**

**ScatterPlots**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary', y='sales',data=df, hue='level of education')
plt.savefig('myplot.jpeg')
```

```python
sns.rugplot(x = 'salary', data = df, height = 0.5)
```

```python
sns.set(style='darkgrid')
sns.displot(data = df, x='salary', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.displot(data=sample_ages, x ='age', rug=True, bins=30, kde=True)
```

**Categorical Plots**

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
```
